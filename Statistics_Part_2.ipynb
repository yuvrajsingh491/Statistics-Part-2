

Q1 What is hypothesis testing in statistics

-Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on a sample of data. It helps you determine whether there is enough statistical evidence to support a specific belief or claim (called a hypothesis) about a population parameter (like a mean or proportion).

Q2 What is the null hypothesis, and how does it differ from the alternative hypothesis?

- The null hypothesis and alternative hypothesis are fundamental concepts in hypothesis testing. Here's a clear breakdown:

üîπ Null Hypothesis (H‚ÇÄ):
It is the default or original assumption about a population.

It assumes no effect, no difference, or no relationship between variables.

The goal of hypothesis testing is often to challenge or disprove the null hypothesis.

Examples:

H‚ÇÄ: The average height of students = 170 cm

H‚ÇÄ: The new drug has no effect compared to the placebo

H‚ÇÄ: There is no correlation between study time and exam scores

Q3 What is the significance level in hypothesis testing, and why is it important?

- The significance level, often denoted by Œ± (alpha), is the threshold for deciding whether to reject the null hypothesis (H‚ÇÄ).

It represents the probability of making a Type I error ‚Äî that is, rejecting the null hypothesis when it is actually true.

üìå Common Significance Levels:
Œ± value	Interpretation
0.05	5% risk of rejecting a true H‚ÇÄ
0.01	1% risk ‚Äî more strict (used in medicine, etc.)
0.10	10% risk ‚Äî more relaxed

Q4 What does a P-value represent in hypothesis testing

- The p-value is the probability of obtaining a test result at least as extreme as the one observed, assuming the null hypothesis is true.

üß† In Simple Terms:
It answers the question:
"If H‚ÇÄ were actually true, how likely is it that I would get this result (or something more extreme) just by random chance?"

Q5 How do you interpret the P-value in hypothesis testing

- The p-value tells you how surprising your sample data is if the null hypothesis (H‚ÇÄ) were true.

üîç Step-by-Step Interpretation:
Start with a null hypothesis (H‚ÇÄ):
A default claim ‚Äî usually no effect, no difference, or status quo.

Calculate the p-value based on your sample data and a chosen statistical test (e.g., t-test, z-test).

Compare the p-value to your significance level (Œ±), commonly set at 0.05.

Q6 What are Type 1 and Type 2 errors in hypothesis testing

- In hypothesis testing, we make decisions based on data ‚Äî but there's always a risk of making a wrong decision. These wrong decisions are called:

üîπ Type I Error (False Positive)
Definition: Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

üí• You think there's an effect, but there isn‚Äôt.
You conclude a new drug works when it actually doesn't.

Equivalent to a false alarm.

üßÆ Probability:
Denoted by Œ± (significance level), often set at 0.05.

Means a 5% chance of wrongly rejecting H‚ÇÄ.

üî∏ Type II Error (False Negative)
Definition: Failing to reject the null hypothesis (H‚ÇÄ) when the alternative hypothesis (H‚ÇÅ) is actually true.

üôà You miss a real effect.
You conclude a new drug doesn‚Äôt work when it actually does.

Equivalent to missing the signal.

üßÆ Probability:
Denoted by Œ≤

(1 - Œ≤) is called power of the test ‚Üí higher power = lower chance of Type II error

Q11 What is the difference between a Z-test and a T-test

- ‚úÖ Z-Test:
Used when the population standard deviation (œÉ) is known.

Best suited for large sample sizes (typically n ‚â• 30).

Assumes the data is normally distributed.

Based on the standard normal distribution (bell curve).

Less variation in results due to thinner tails.

Example: Testing if the average height is 170 cm when œÉ is known and sample size is large.

‚úÖ T-Test:
Used when the population standard deviation is unknown.

Suitable for small sample sizes (typically n < 30), though it can be used for larger samples too.

Assumes the sample data is approximately normal.

Based on the t-distribution, which has fatter tails to handle more variability.

More conservative ‚Äî accounts for extra uncertainty in small samples.

Example: Testing if the average test score is 75 when œÉ is unknown and the sample size is 15.

Q21 What are the different types of ANOVA tests

-1. One-Way ANOVA
Purpose: Compares the means of three or more independent groups based on one categorical independent variable.

Example: Comparing test scores of students across three different schools.

‚úÖ Use when:

One independent variable (factor)

One dependent variable (continuous)

üî¢ 2. Two-Way ANOVA (Factorial ANOVA)
Purpose: Tests the effect of two independent variables on one continuous outcome, and also checks for interaction between them.

Example: Studying the effect of diet type and exercise level on weight loss.

‚úÖ It tells you:

Main effect of factor A (e.g., diet)

Main effect of factor B (e.g., exercise)

Interaction effect A √ó B (do the effects depend on each other?)

üîÅ 3. Repeated Measures ANOVA
Purpose: Used when the same subjects are measured multiple times (within-subject design).

Example: Measuring blood pressure of patients before, during, and after treatment.

‚úÖ Use when:

You have dependent (related) samples

Reduces variability since the same subjects are used across treatments

üîÅüî¢ 4. Two-Way Repeated Measures ANOVA
Purpose: Extension of repeated measures ANOVA with two within-subject factors (or one within, one between).

Example: Testing performance across multiple tasks (factor 1) over time (factor 2) in the same group of participants.

üß™ 5. MANOVA (Multivariate Analysis of Variance)
Purpose: Tests multiple dependent variables simultaneously.

Example: Testing how teaching method affects both test scores and student satisfaction.

‚úÖ Useful when:

You have more than one quantitative outcome

You want to control Type I error from running multiple ANOVAs

```python
#Q3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
```

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    float: P(X = x)
    """
    # Check that probability p is valid
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    # Return the PDF based on value of x
    if x == 0:
        return 1 - p
    elif x == 1:
        return p
    else:
        return 0  # For any x not 0 or 1


# Test the function
p = 0.7  # Probability of success

# Test outcomes
print("P(X = 0):", bernoulli_pdf(0, p))  # Expected: 0.3
print("P(X = 1):", bernoulli_pdf(1, p))  # Expected: 0.7
print("P(X = 2):", bernoulli_pdf(2, p))  # Expected: 0
```

```python
#Q9 Write a Python function to calculate Z-scores from a dataset and plot them
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.

    Parameters:
    data (array-like): Input numerical dataset

    Returns:
    np.ndarray: Array of Z-scores
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Generate a random dataset (normal distribution)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(10, 4))
plt.plot(z_scores, marker='o', linestyle='', color='purple', alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.axhline(y=1, color='green', linestyle='--', linewidth=1, label='Z = ¬±1')
plt.axhline(y=-1, color='green', linestyle='--', linewidth=1)
plt.axhline(y=2, color='orange', linestyle='--', linewidth=1, label='Z = ¬±2')
plt.axhline(y=-2, color='orange', linestyle='--', linewidth=1)
plt.title("Z-scores of Dataset")
plt.xlabel("Data Index")
plt.ylabel("Z-score")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#Q15 Implement hypothesis testing using Z-statistics for a sample dataset
```

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a two-tailed Z-test for a sample mean.

    Parameters:
    - sample: list or array of sample data
    - population_mean: hypothesized population mean (Œº‚ÇÄ)
    - population_std: population standard deviation (œÉ)
    - alpha: significance level (default = 0.05)

    Returns:
    - Z-statistic, p-value, and test result
    """
    sample_mean = np.mean(sample)
    n = len(sample)

    # Calculate Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Decision
    if p_value < alpha:
        decision = "Reject null hypothesis (H‚ÇÄ)"
    else:
        decision = "Fail to reject null hypothesis (H‚ÇÄ)"

    # Print result
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Conclusion: {decision}")

# üîç Example usage
sample_data = [52, 55, 49, 50, 54, 48, 51, 53, 50, 52]
population_mean = 50     # Œº‚ÇÄ
population_std = 3       # known œÉ
alpha = 0.05             # significance level

z_test(sample_data, population_mean, population_std, alpha)
```

```python
#Q16 Create a confidence interval for a dataset using Python and interpret the resultE
```

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default = 0.95)

    Returns:
    - Tuple: (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t critical value

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval = ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# üîç Example usage
sample_data = [45, 48, 50, 52, 47, 49, 46, 51, 53, 50]
confidence_interval(sample_data, confidence=0.95)
```

```python
#Q17 Generate data from a normal distribution, then calculate and interpret the confidence interval for its meanE
```

```python
import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu = 100            # True population mean
sigma = 15          # True population standard deviation
n = 30              # Sample size

# Generate sample
sample = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate the confidence interval (CI)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
standard_error = sample_std / np.sqrt(n)

# Get the t-critical value for 95% confidence
confidence = 0.95
t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

# Compute margin of error and CI
margin_of_error = t_crit * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 3: Output the result
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
# distribution.
```

```python
import numpy as np
from scipy.stats import skew, kurtosis

def compute_skewness_kurtosis(n=1000, mean=0, std=1):
    """
    Generates data from a normal distribution and calculates skewness and kurtosis.

    Parameters:
    - n: Number of data points
    - mean: Mean of the normal distribution
    - std: Standard deviation of the normal distribution

    Returns:
    - Tuple of (skewness, kurtosis)
    """
    # Generate data
    data = np.random.normal(loc=mean, scale=std, size=n)

    # Calculate skewness and kurtosis
    skewness = skew(data)
    excess_kurtosis = kurtosis(data)  # excess kurtosis: normal ‚âà 0

    # Print results
    print(f"Skewness: {skewness:.4f} (‚âà 0 for symmetric)")
    print(f"Excess Kurtosis: {excess_kurtosis:.4f} (‚âà 0 for normal distribution)")

    return skewness, excess_kurtosis

# üîç Run the function
compute_skewness_kurtosis()
```

```python
import numpy as np
from scipy.stats import norm

def z_test_sample_mean(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for comparing a sample mean to a known population mean.

    Parameters:
    - sample: list or array of sample values
    - population_mean: hypothesized population mean (Œº‚ÇÄ)
    - population_std: known population standard deviation (œÉ)
    - alpha: significance level (default = 0.05)

    Returns:
    - Z-score, p-value, and decision interpretation
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # two-tailed test

    # Interpretation
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Population Mean (Œº‚ÇÄ): {population_mean}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject the null hypothesis at Œ± = {alpha}")
        print("Interpretation: The sample mean is significantly different from the population mean.")
    else:
        print(f"Result: Fail to reject the null hypothesis at Œ± = {alpha}")
        print("Interpretation: No significant difference between sample and population mean.")

# üîç Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=52, scale=10, size=50)  # sample with mean near 52
population_mean = 50   # known population mean
population_std = 10    # known population standard deviation
alpha = 0.05           # significance level

z_test_sample_mean(sample_data, population_mean, population_std, alpha)
```

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Performs a one-sample Z-test comparing the sample mean to the population mean.

    Parameters:
    - sample: array-like sample data
    - population_mean: known population mean (Œº‚ÇÄ)
    - population_std: known population standard deviation (œÉ)
    - alpha: significance level (default = 0.05)

    Returns:
    - Prints Z-score, p-value, and decision
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    standard_error = population_std / np.sqrt(n)
    z_score = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed

    # Print results
    print(f"Sample Size: {n}")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Population Mean (Œº‚ÇÄ): {population_mean}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject H‚ÇÄ at Œ± = {alpha}. Sample mean is significantly different.")
    else:
        print(f"Result: Fail to reject H‚ÇÄ at Œ± = {alpha}. No significant difference found.")

# üîç Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=40)  # Simulated data
population_mean = 50
population_std = 10

one_sample_z_test(sample_data, population_mean, population_std)
```

```python
#Q4 Perform a two-tailed Z-test using Python and visualize the decision region on a plotE
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_visual(sample, population_mean, population_std, alpha=0.05):
    # Step 1: Calculate statistics
    n = len(sample)
    sample_mean = np.mean(sample)
    se = population_std / np.sqrt(n)
    z_score = (sample_mean - population_mean) / se
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Step 2: Print test results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Step 3: Plot the normal distribution and critical regions
    z_vals = np.linspace(-4, 4, 1000)
    pdf_vals = norm.pdf(z_vals)

    z_critical = norm.ppf(1 - alpha/2)

    plt.figure(figsize=(10, 5))
    plt.plot(z_vals, pdf_vals, label="Standard Normal Distribution", color='blue')

    # Fill rejection regions
    plt.fill_between(z_vals, 0, pdf_vals, where=(z_vals <= -z_critical), color='red', alpha=0.4, label="Rejection Region")
    plt.fill_between(z_vals, 0, pdf_vals, where=(z_vals >= z_critical), color='red', alpha=0.4)

    # Plot the Z-score
    plt.axvline(z_score, color='green', linestyle='--', label=f"Z = {z_score:.2f}")
    plt.axvline(-z_critical, color='black', linestyle='--', label=f"¬±Z critical = ¬±{z_critical:.2f}")
    plt.axvline(z_critical, color='black', linestyle='--')

    # Labels and legend
    plt.title("Two-Tailed Z-Test: Decision Region")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    # Step 4: Conclusion
    if abs(z_score) > z_critical:
        print(f"Result: Reject H‚ÇÄ at Œ± = {alpha}. Sample mean is significantly different.")
    else:
        print(f"Result: Fail to reject H‚ÇÄ at Œ± = {alpha}. No significant difference found.")

# üîç Example usage
np.random.seed(1)
sample = np.random.normal(loc=53, scale=10, size=40)
population_mean = 50
population_std = 10
alpha = 0.05

two_tailed_z_test_visual(sample, population_mean, population_std, alpha)
```

```python
#Q5  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testingE
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05):
    """
    Visualizes Type I and Type II errors for a one-tailed Z-test (right-tailed).

    Parameters:
    - mu0: mean under null hypothesis H‚ÇÄ
    - mu1: mean under alternative hypothesis H‚ÇÅ
    - sigma: population standard deviation
    - n: sample size
    - alpha: significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for rejecting H‚ÇÄ (right-tailed test)
    z_alpha = norm.ppf(1 - alpha)
    critical_value = mu0 + z_alpha * se

    # Compute Œ≤: Type II error
    z_beta = (critical_value - mu1) / se
    beta = norm.cdf(z_beta)
    power = 1 - beta

    # Plot distributions under H‚ÇÄ and H‚ÇÅ
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    h0_pdf = norm.pdf(x, mu0, se)
    h1_pdf = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, h0_pdf, label="H‚ÇÄ: N(Œº‚ÇÄ, œÉ¬≤)", color='blue')
    plt.plot(x, h1_pdf, label="H‚ÇÅ: N(Œº‚ÇÅ, œÉ¬≤)", color='orange')

    # Shade Type I error region (alpha)
    plt.fill_between(x, 0, h0_pdf, where=(x > critical_value), color='red', alpha=0.3, label='Type I Error (Œ±)')

    # Shade Type II error region (beta)
    plt.fill_between(x, 0, h1_pdf, where=(x < critical_value), color='green', alpha=0.3, label='Type II Error (Œ≤)')

    # Vertical line at critical value
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # Print results
    print(f"Critical Value: {critical_value:.2f}")
    print(f"Z (Œ±): {z_alpha:.2f} ‚Üí Œ± = {alpha}")
    print(f"Z (Œ≤): {z_beta:.2f} ‚Üí Œ≤ = {beta:.4f}")
    print(f"Power of the Test: {power:.4f}")

# üîç Example usage
visualize_type1_type2(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05)
```

```python
#Q6 Write a Python program to perform an independent T-test and interpret the results
```

```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(group1, group2, alpha=0.05):
    """
    Perform an independent two-sample T-test.

    Parameters:
    - group1: First group of data (array-like)
    - group2: Second group of data (array-like)
    - alpha: Significance level (default = 0.05)
    """
    # Perform the T-test (assume unequal variances by default ‚Äî Welch‚Äôs t-test)
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)

    # Print the results
    print(f"Group 1 Mean: {np.mean(group1):.2f}")
    print(f"Group 2 Mean: {np.mean(group2):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject H‚ÇÄ at Œ± = {alpha}. The means are significantly different.")
    else:
        print(f"Result: Fail to reject H‚ÇÄ at Œ± = {alpha}. No significant difference in means.")

# üîç Example usage
np.random.seed(42)
group1 = np.random.normal(loc=55, scale=10, size=30)  # Mean ~ 55
group2 = np.random.normal(loc=60, scale=10, size=30)  # Mean ~ 60

independent_t_test(group1, group2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test_with_plot(before, after, alpha=0.05):
    """
    Perform a paired T-test and visualize the result.

    Parameters:
    - before: array-like (before treatment)
    - after: array-like (after treatment)
    - alpha: significance level (default 0.05)
    """

    # Paired T-test
    t_stat, p_value = ttest_rel(before, after)
    differences = np.array(after) - np.array(before)

    # Print results
    print(f"Mean (Before): {np.mean(before):.2f}")
    print(f"Mean (After): {np.mean(after):.2f}")
    print(f"Mean Difference: {np.mean(differences):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject H‚ÇÄ at Œ± = {alpha}. The change is statistically significant.")
    else:
        print(f"Result: Fail to reject H‚ÇÄ at Œ± = {alpha}. No significant change detected.")

    # Visualization
    plt.figure(figsize=(10, 5))

    # 1. Line plot to show paired values
    plt.subplot(1, 2, 1)
    for i in range(len(before)):
        plt.plot([1, 2], [before[i], after[i]], marker='o', color='gray')
    plt.xticks([1, 2], ['Before', 'After'])
    plt.title("Paired Observations")
    plt.ylabel("Value")
    plt.grid(True, linestyle='--', alpha=0.5)

    # 2. Histogram of differences
    plt.subplot(1, 2, 2)
    plt.hist(differences, bins=10, color='lightgreen', edgecolor='black')
    plt.axvline(np.mean(differences), color='red', linestyle='--', label='Mean Difference')
    plt.title("Histogram of Differences (After - Before)")
    plt.xlabel("Difference")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# üîç Example usage
np.random.seed(1)
before = np.random.normal(loc=65, scale=5, size=30)
after = before + np.random.normal(loc=-2, scale=3, size=30)  # some decrease expected

paired_t_test_with_plot(before, after)
```

```python
import numpy as np
from scipy.stats import norm, t

def compare_z_test_and_t_test(mu=100, sigma=15, sample_size=30, alpha=0.05):
    # Step 1: Simulate sample data
    np.random.seed(42)
    sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample std dev
    se = sample_std / np.sqrt(sample_size)

    # Step 2: Hypothesized population mean
    population_mean = 100  # H0: mu = 100

    # Step 3: Z-test (assuming known sigma)
    z_se = sigma / np.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / z_se
    z_p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Step 4: T-test (sigma unknown)
    t_score = (sample_mean - population_mean) / se
    t_p_value = 2 * (1 - t.cdf(abs(t_score), df=sample_size - 1))

    # Step 5: Output comparison
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Std Dev: {sample_std:.2f}")
    print(f"Sample Size: {sample_size}")
    print(f"Population Mean (H‚ÇÄ): {population_mean}\n")

    print("üîπ Z-test (œÉ known):")
    print(f"  Z-score: {z_score:.4f}")
    print(f"  P-value: {z_p_value:.4f}")
    if z_p_value < alpha:
        print("  ‚û§ Reject H‚ÇÄ (significant difference)")
    else:
        print("  ‚û§ Fail to reject H‚ÇÄ (no significant difference)")

    print("\nüîπ T-test (œÉ unknown):")
    print(f"  T-score: {t_score:.4f}")
    print(f"  P-value: {t_p_value:.4f}")
    if t_p_value < alpha:
        print("  ‚û§ Reject H‚ÇÄ (significant difference)")
    else:
        print("  ‚û§ Fail to reject H‚ÇÄ (no significant difference)")

# üîç Run the comparison
compare_z_test_and_t_test()
```

```python
#Q9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance
```

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - data: array-like sample data
    - confidence: desired confidence level (default = 0.95)

    Returns:
    - (lower_bound, upper_bound) of the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # Standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)  # Critical t-value

    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Level: {int(confidence * 100)}%")
    print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")

    return lower, upper

# üîç Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)
confidence_interval(sample_data)
```

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at the given confidence level.

    Parameters:
    - data: array-like sample data
    - confidence: confidence level (default = 0.95 for 95%)

    Returns:
    - Margin of error (float)
    """
    n = len(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation
    standard_error = sample_std / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_error = t_critical * standard_error

    print(f"Sample Size: {n}")
    print(f"Sample Std Dev: {sample_std:.2f}")
    print(f"Confidence Level: {int(confidence * 100)}%")
    print(f"Margin of Error: ¬±{margin_error:.2f}")

    return margin_error

# üîç Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=75, scale=10, size=40)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
#Q11  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
```

```python
def bayes_theorem(prior, likelihood, false_positive, prior_not=None):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(D), prior probability of disease/hypothesis
    - likelihood: P(Pos|D), probability of positive test given disease
    - false_positive: P(Pos|¬¨D), probability of positive test given no disease
    - prior_not: P(¬¨D), optional; if None, calculated as (1 - prior)

    Returns:
    - posterior: P(D|Pos), updated belief after seeing evidence
    """
    if prior_not is None:
        prior_not = 1 - prior

    # Total probability of a positive result
    marginal = (likelihood * prior) + (false_positive * prior_not)

    # Bayes' Theorem
    posterior = (likelihood * prior) / marginal

    print(f"Prior (P(D)): {prior:.4f}")
    print(f"Likelihood (P(Pos|D)): {likelihood:.4f}")
    print(f"False Positive Rate (P(Pos|¬¨D)): {false_positive:.4f}")
    print(f"Marginal P(Pos): {marginal:.4f}")
    print(f"Posterior (P(D|Pos)): {posterior:.4f}")

    return posterior

# üîç Example usage
bayes_theorem(prior=0.01, likelihood=0.99, false_positive=0.05)
```

```python
#Q12  Perform a Chi-square test for independence between two categorical variables in PythonE
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def chi_square_independence_test():
    """
    Performs a Chi-square test for independence between two categorical variables.
    """
    # Example: Contingency table (Observed Frequencies)
    #           Likes | Dislikes
    # Male        20   |    10
    # Female      30   |    40

    data = np.array([[20, 10],
                     [30, 40]])

    categories = ['Male', 'Female']
    responses = ['Likes', 'Dislikes']

    df = pd.DataFrame(data, index=categories, columns=responses)
    print("Contingency Table (Observed):\n")
    print(df, "\n")

    # Perform Chi-square test
    chi2, p, dof, expected = chi2_contingency(data)

    print(f"Chi-square Statistic: {chi2:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print(f"P-value: {p:.4f}")
    print("\nExpected Frequencies:\n")
    print(pd.DataFrame(expected, index=categories, columns=responses))

    # Interpretation
    alpha = 0.05
    if p < alpha:
        print(f"\nResult: Reject H‚ÇÄ (variables are NOT independent at Œ± = {alpha})")
    else:
        print(f"\nResult: Fail to reject H‚ÇÄ (variables are independent at Œ± = {alpha})")

# üîç Run the test
chi_square_independence_test()
```

```python
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed, expected_probs=None):
    """
    Performs a Chi-square goodness-of-fit test.

    Parameters:
    - observed: array of observed frequencies
    - expected_probs: list of expected probabilities (must sum to 1). If None, assumes uniform distribution.

    Returns:
    - Chi-square statistic and p-value
    """
    observed = np.array(observed)
    n = observed.sum()

    if expected_probs is None:
        expected = np.full_like(observed, fill_value=n / len(observed))
    else:
        expected_probs = np.array(expected_probs)
        expected = expected_probs * n

    chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

    print("Observed:", observed)
    print("Expected:", np.round(expected, 2))
    print(f"\nChi-square Statistic: {chi_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print(f"Result: Reject H‚ÇÄ ‚Üí Observed data does NOT fit expected distribution (Œ± = {alpha})")
    else:
        print(f"Result: Fail to reject H‚ÇÄ ‚Üí Observed data fits expected distribution (Œ± = {alpha})")

# üîç Example 1: Uniform distribution expected
observed_freq = [18, 22, 20, 19, 21]
goodness_of_fit_test(observed_freq)

# üîç Example 2: Custom expected distribution (e.g., dice loaded toward 6)
# expected_probs = [1/12, 1/12, 1/6, 1/6, 1/6, 1/3]
# observed_freq = [8, 9, 15, 17, 18, 33]
# goodness_of_fit_test(observed_freq, expected_probs)
```
